```python
F=artist_F.to_numpy()
F_likes=np.count_nonzero(F==1)
F_dislikes=np.count_nonzero(F==0)

C=artist_C.to_numpy()
C_likes=np.count_nonzero(C==1)
C_dislikes=np.count_nonzero(C==0)

B=artist_B.to_numpy()
B_likes=np.count_nonzero(B==1)
B_dislikes=np.count_nonzero(B==0)

J=artist_J.to_numpy()
J_likes=np.count_nonzero(J==1)
J_dislikes=np.count_nonzero(J==0)

D=artist_D.to_numpy()
D_likes=np.count_nonzero(D==1)
D_dislikes=np.count_nonzero(D==0)
```

```python
raw_data = {'artist_name': ['F', 'C', 'B','J','D'],
        'likes': [F_likes,C_likes, B_likes,J_likes,D_likes],
        'dislikes': [F_dislikes, C_dislikes, B_dislikes,J_dislikes,D_dislikes],
        }
df = pd.DataFrame(raw_data, columns = ['artist_name', 'likes', 'dislikes'])
df
```
